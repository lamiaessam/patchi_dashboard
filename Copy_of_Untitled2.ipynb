```python
import pandas as pd
import numpy as np
```

```python
excel_sheet ='/content/Data model.xlsx'
sheets_location =pd.read_excel(excel_sheet,sheet_name=None)
```

```python
sales = sheets_location['factSalesTable']
customer = sheets_location['dimCustomerTable']
product = sheets_location['dimProductTable']
product_category = sheets_location['dimProductCategoryTable']
product_subcategory = sheets_location['dimProductSubcategoryTable']
calender = sheets_location['dimDateTable']
region = sheets_location['dimRegionTable']
```

```python
df_merge = pd.merge(sales,customer,on='CustomerKey',how='left')
```

```python
df_merge =pd.merge(df_merge,region,on='GeographyKey',how='left')
```

```python
df_merge = pd.merge(df_merge,product,on='ProductKey',how='left')
```

```python
df_merge = pd.merge(df_merge,product_subcategory,on='ProductSubcategoryKey',how='left')
```

```python
df_merge = pd.merge(df_merge,product_category,on='ProductCategoryKey',how='left')
```

```python
df_merge = pd.merge(df_merge,calender,left_on='OrderDateKey',right_on='DateKey',how='left')
```
